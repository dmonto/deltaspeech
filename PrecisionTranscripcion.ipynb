```
#@title Dar play a la izda
!pip install jiwer
import os
import pandas as pd
import math
from jiwer import wer

def precisiontrans(fil):
  df = pd.read_csv(fil, sep=";")
  for fich in df["Fichero"].unique():
    print("_____________________________________________")
    print("Resumen para {}".format(fich))
    summ = pd.DataFrame(columns=["Fichero", "Sistema", "Texto"])
    dfuno = df[df["Fichero"] == fich]
    human = dfuno[dfuno["Sistema"] == "Human"]
    if len(human) == 0 or type(human["Texto"]) is not pd.core.series.Series or type(human["Texto"].iloc[0]) is not str:
        print("\tFalta Transcripción Humana!")
        continue
    ground_truth = human["Texto"]
    for tool in dfuno.iterrows():
        hypothesis = tool[1]["Texto"]
        if type(hypothesis) is str and len(hypothesis):
            error = wer(ground_truth.values, hypothesis)
            summ = summ.append({"Acc": (1 - error) * 100, "Tool": tool[1]["Sistema"], "Trans": hypothesis}, ignore_index=True)
    for ind in summ.sort_values("Acc", ascending=False).index:
        print("\tPrecisión para '{}' vs '{}': \t{:.2f}%".format(summ["Tool"][ind], "Human", summ["Acc"][ind]))
        if summ["Acc"][ind] < .1:
            print("Algo esta muy mal con esta transcripción:")
            print(summ["Trans"][ind])

directory = "."
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.endswith(".csv") and filename.startswith("Trans") :
        fil = os.path.join(directory, filename)
        print("=====================================")
        print("Leyendo {}".format(fil))
        precisiontrans(fil)
```
